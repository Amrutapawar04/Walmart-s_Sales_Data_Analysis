```python
import pandas as pd
```

```python
df = pd.read_csv("Walmart.csv")
```

```python
print("First 5 rows:")
print(df.head())
```

```python
print("\nDataset Info:")
print(df.info())
```

```python
print("\nStatistical Summary:")
print(df.describe())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
```

```python
df.rename(columns={
    'Weekly_Sales': 'WeeklySales',
    'Holiday_Flag': 'HolidayFlag',
    'Fuel_Price': 'FuelPrice'
}, inplace=True)

```

```python
df['HolidayFlag'] = df['HolidayFlag'].astype('category')
```

```python
df.sort_values(by=['Store', 'Date'], inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
# Step 11: Save Cleaned Dataset
df.to_csv("C:/Users/YourName/Desktop/Walmart_Cleaned.csv", index=False)
print("\nData Cleaning Completed Successfully!")

```

Now Vizualization Part Start

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1️⃣ SALES TREND OVER TIME

plt.figure(figsize=(10,5))
df.groupby('Date')['Weekly_Sales'].sum().plot()
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Weekly Sales")
plt.show()
```

```python
#Sales increase at year end.
```

```python
#2️⃣ STORE PERFORMANCE

plt.figure(figsize=(10,5))
df.groupby('Store')['Weekly_Sales'].sum().plot(kind='bar')
plt.title("Store Wise Sales")
plt.show()
```

```python
#Store 20 highest sales.
```

```python
# 3️⃣ HOLIDAY IMPACT

sns.barplot(x='Holiday_Flag', y='Weekly_Sales', data=df)
plt.title("Holiday vs Normal Sales")
plt.show()
```

```python
#Holiday weeks higher sales.
```

```python
#4️⃣ SCATTER — FUEL PRICE vs SALES

plt.scatter(df['Fuel_Price'], df['Weekly_Sales'])
plt.xlabel("Fuel Price")
plt.ylabel("Weekly Sales")
plt.title("Fuel Price vs Sales")
plt.show()
```
